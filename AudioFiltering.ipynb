```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.io.wavfile
from IPython.display import Audio
from scipy.signal import gaussian
```

```python
# Load the Audio 
Omega, f = scipy.io.wavfile.read('recording.wav')
Audio(f, rate=Omega)
```

```python
N = len(f)
L = N/Omega
t = np.arange(0, N) * L/N

omega = np.fft.fftshift(np.arange(-N/2, N/2)) / L
```

```python
plt.plot(t, f)
plt.title('Sound')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
```

```python
F = np.fft.fft(f)
plt.plot(omega, abs(F))
plt.title('Frequency vs Modulus')
plt.xlabel('Frequency')
plt.ylabel('Modulus of Fourier coefficients')
```

```python
THRESHOLD_FREQUENCY = 2350

threshold = THRESHOLD_FREQUENCY

size = len(F)
g = F

for i in range(size):
  if (abs(omega[i]) > threshold):
    g[i] = 0

plt.plot(omega, abs(g))
plt.title('Filtered over threshold')
plt.xlabel('Frequency(Hz)')
plt.ylabel('Modulus of Fourier coefficients')
```

```python
f_filtered = np.fft.ifft(g)

plt.plot(t, f_filtered.real)
plt.title('Final Sound Clip')
plt.xlabel('Time(s)')
plt.ylabel('Modulus')
```

```python
Audio(f_filtered.real, rate=Omega)
```
